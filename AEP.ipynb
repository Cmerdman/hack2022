Data Preprocessing

```python
X = datA.drop(columns = ['Turbine', 'Date & Time'],axis=1)
Y = datA['Turbine']
```

Normalize Data

```python
normalized_X=(X-X.mean())/X.std()
```

```python
normalized_X = normalized_X.dropna(axis=1, how='all')
normalized_X.head(15)
```

```python
normalized_X.dtypes
```

```python
fa = FactorAnalyzer(rotation = None,impute = "drop",n_factors=normalized_X.shape[1])
fa.fit(normalized_X)
ev,_ = fa.get_eigenvalues()

plt.scatter(range(1,normalized_X.shape[1]+1),ev)
plt.plot(range(1,normalized_X.shape[1]+1),ev)
plt.title('Scree Plot')
plt.xlabel('Factors')
plt.ylabel('Variance Explained')
plt.grid()
```
